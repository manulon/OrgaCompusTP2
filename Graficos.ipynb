```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import random as rn
%matplotlib inline
plt.style.use('seaborn')
```

```python
# Generamos los digitos de las matrices

numeros = []
for i in range( (16*16) + (16*16) ):
    numero = rn.randint(0,9)
    numeros.append(numero)

    
for i in numeros:
    print(i,"",end='')
```

```python
misses_instr_dm = [14,14,14,14]
total_instr = [524,150618,9090282,573248298]
miss_rate_instr_dm = [( misses_instr_dm[0]/total_instr[0] ),( misses_instr_dm[1]/total_instr[1] ) \
                     ,( misses_instr_dm[2]/total_instr[2] ),( misses_instr_dm[3]/total_instr[3] )]

misses_datos_dm = [6,196,18939,17149461]
total_datos = [244,74878,4538830,286525198]
miss_rate_datos_dm = [( misses_datos_dm[0]/total_datos[0] ),( misses_datos_dm[1]/total_datos[1] )\
                     ,( misses_datos_dm[2]/total_datos[2]),( misses_datos_dm[3]/total_datos[3] )]

# Esta en escala log en base 2
dimension_matriz = [1,4,6,8]
```

```python
data1 = misses_instr_dm
data2 = miss_rate_instr_dm

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Dimension de la matriz (escala log en base 2)')
ax1.set_ylabel('Instruction miss count ', color=color)
ax1.plot(dimension_matriz, data1, color=color, marker = 'o', label = 'Instruction miss count')
plt.legend(loc = 'lower left')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx() 

color = 'tab:blue'
ax2.set_ylabel('Instruccion miss rate', color=color)  
ax2.set_yscale('log')
ax2.plot(dimension_matriz, data2, color=color, marker = 'o', label = 'Instruction miss rate')
ax2.tick_params(axis='y', labelcolor=color)
fig.tight_layout()
plt.legend(loc = 'upper right')
plt.show()
```

```python
data1 = misses_datos_dm
data2 = miss_rate_datos_dm

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Dimension de la matriz (escala log en base 2)')
ax1.set_ylabel('Data miss count ', color=color)
ax1.set_yscale('log')
ax1.plot(dimension_matriz, data1, color=color, marker = 'o', label = 'Data miss count')
plt.legend(loc = 'lower right')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx() 

color = 'tab:blue'
ax2.set_ylabel('Data miss rate', color=color)  
ax2.plot(dimension_matriz, data2, color=color, marker = 'o', label = 'Data miss rate')
ax2.tick_params(axis='y', labelcolor=color)
fig.tight_layout()
plt.legend(loc = 'upper right')
plt.show()    
```
